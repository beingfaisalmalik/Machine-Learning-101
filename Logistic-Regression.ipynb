# Logistic Regression Classification

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = sns.load_dataset('iris')
df.head()
```

```python
#for Checking all the available classes of predictions
df['species'].unique()
```

```python
df.isnull().sum()
```

```python
#for removing the choice of the third class
df = df[df['species'] != 'setosa']
df.head()
```

```python
#converting the resultant class into 0 and 1 for classification

df['species'] = df['species'].map({'versicolor':0,'virginica':1})
df.head()
```

```python
X = df.iloc[:,:-1]
X
```

```python
Y = df.loc[:,'species']
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_Test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25 , random_state=42) 
```

```python
X_train
```

```python
print(np.shape(X_Test))
X_Test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.linear_model import LogisticRegression
Classifier = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {'penalty':['l1','l2','elasticnet'],'C':[i for i in range(1,101)]}
```

```python
classifier_regressor = GridSearchCV(Classifier,parameter,scoring='accuracy',cv=5)
```

```python
classifier_regressor.fit(X_train,Y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```

```python
Y_predict = classifier_regressor.predict(X_Test)
```

```python
print(Y_predict)
print(Y_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(Y_test,Y_predict)
```
